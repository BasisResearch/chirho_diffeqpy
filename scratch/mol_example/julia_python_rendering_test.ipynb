```python
from pygments import highlight
from pygments.lexers import get_lexer_by_name
from pygments.formatters import HtmlFormatter
from IPython.core.display import HTML
```

```python
# Define Julia code as a string
julia_code = """
x = [1, 2, 3]
mean(x)
"""

# Use the Julia lexer for syntax highlighting
lexer = get_lexer_by_name("julia")

# Use an appropriate formatter with full HTML styling for colors
formatter = HtmlFormatter(style="colorful", full=True)

# Highlight the code
highlighted_code = highlight(julia_code, lexer, formatter)

# Display the syntax-highlighted Julia code with HTML
HTML(highlighted_code)
```
